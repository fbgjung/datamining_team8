```python
from konlpy.tag import Okt
okt = Okt()
```

```python
word_pos_list = []
f = open('news_title_pos.txt','r')
f1 = open('news_title_neg.txt','r')
g_string = ",".join(f)
b_string = ",".join(f1)
g_nouns = okt.nouns(g_string)
b_nouns = okt.nouns(b_string)
g_word = [x for x in g_nouns if len(x) >= 2]
b_word = [x for x in b_nouns if len(x) >= 2]
stopwords = ["삼성","전자","하이닉스","디스플레이","코스피","주가","반도체","배터리","매수","사업","종목","외국인","기관","화학"]
g_word = [i for i in g_word if i not in stopwords]
b_word = [i for i in b_word if i not in stopwords]
```

```python
from collections import Counter
g_count = Counter(g_word)
b_count = Counter(b_word)
print('<긍정>')
print(g_count)
print('<부정>')
print(b_count)
```
