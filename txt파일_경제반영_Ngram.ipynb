```python
from konlpy.tag import Okt
okt = Okt()
```

```python
word_pos_list = []
f = open('긍정뉴스타이틀.txt','r')
f1 = open('부정뉴스 타이틀.txt','r')
g_string = ",".join(f)
b_string = ",".join(f1)
g_nouns = okt.nouns(g_string)
b_nouns = okt.nouns(b_string)
g_word = [x for x in g_nouns if len(x) >= 2]
b_word = [x for x in b_nouns if len(x) >= 2]
g_word_n = []
b_word_n = []
for i in range(len(g_word) - 1):
  n_gram = (g_word[i]+g_word[i + 1])
  g_word_n.append(n_gram)
for i in range(len(b_word) - 1):
  n_gram = (b_word[i]+b_word[i + 1])
  b_word_n.append(n_gram)
#긍정기사와 부정기사를 각각에 나온 단어들을 앞뒤로 두 개의 단어끼리 묶은 리스트로 변동(n-gram)
stopwords = ["삼성전자","전자하이닉스","상위종목","삼성디스플레이","전자삼성","경제라이프","삼성현대차","하이닉스초반"] 
g_word_n = [i for i in g_word_n if i not in stopwords]
b_word_n = [i for i in b_word_n if i not in stopwords]
# 의미 없는 단어들 제거
```

```python
from collections import Counter
g_count = Counter(g_word_n)
b_count = Counter(b_word_n)
print("n_gram 적용")
print('<긍정>')
print(g_count)
print('<부정>')
print(b_count)
#긍정과 부정 기사에서 각각 나온 나오는 단어들을 빈도수와 함께 정리
```

```python
g_word = [x for x in g_nouns if len(x) >= 2]
b_word = [x for x in b_nouns if len(x) >= 2]
g_word1 = [i for i in g_word if i not in b_word]
b_word1= [i for i in b_word if i not in g_word]
from collections import Counter
g_count = Counter(g_word1)
b_count = Counter(b_word1)
print("겹치지 않는 단어")
print('<긍정>')
print(g_count)
print('<부정>')
print(b_count)
# 긍정기사에서만 나오거나 부정기사에서만 나온 단어들을 빈도수와 함께 정리
```
